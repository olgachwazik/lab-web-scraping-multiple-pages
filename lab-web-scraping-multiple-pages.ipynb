```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

Find other lists of hot songs on the internet and scrape them too: having a bigger pool of songs will be awesome!

### Toplist from Polish Radio Trójka

```python
r = requests.get('https://trojka.polskieradio.pl/notowania').content
soup = BeautifulSoup(r, 'html.parser')
print(soup.prettify())
```

```python
#getting all the artists
artists = soup.find_all("h3", attrs={"class": "VotingList_songArtist__a4XJv"})
artists
```

```python
artists_names = [artist.get_text(strip=True) for artist in artists] #creating the list of all the artist names
artists_names
```

```python
# getting all the songs 
songs = soup.find_all("h4", attrs={"class": "VotingList_songTitle__hY9pS"})
songs
```

```python
songs_titles = [song.get_text(strip=True) for song in songs] #creating the list of all the titles
songs_titles
```

```python
trojka = pd.DataFrame((zip(artists_names, songs_titles)), columns=["artist name", "song name"]) #creating dataframe with all the artists and songs
trojka
```

### AFI's 100 Years ... 100 Songs scrolled from Wikipedia

```python
r = requests.get('https://en.wikipedia.org/wiki/AFI%27s_100_Years...100_Songs').content
soup = BeautifulSoup(r, 'html.parser')
print(soup.prettify())
```

```python
table = soup.find('table', {'class': 'wikitable'})
```

```python
#looping through the table to get the song titles and artists names
l_songs = []
l_artists = []
for row in table.find_all('tr')[1:]:
    columns = row.find_all('td')
    songs = columns[1].text.strip()
    artists = columns[4].text.strip()
    l_songs.append(songs)
    l_artists.append(artists)
```

```python
afi = pd.DataFrame((zip(l_artists, l_songs)), columns=["artist name", "song name"]) #creating dataframe with all the artists and songs
afi
```
